```python
df = pd.read_json(newFile,lines=True)
```

```python
print("The state with the most confirmed cases: ","\n",df.loc[df.Confirmed.idxmax()])
```

```python
print("The state with the least confirmed cases:","\n",df.loc[df.Confirmed.idxmin()])
```
